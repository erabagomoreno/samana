```python
data_class = Mock1Data()
model = Mock1Model
preset_model_name = 'WDM'

# set the priors on dark matter parameters passed into pyHalo https://github.com/dangilman/pyHalo
kwargs_sample_realization = {'log10_sigma_sub': ['UNIFORM', -2.5, -1.0], 
                            'log_mc': ['UNIFORM', 4.0, 4.1]
                            }

kwargs_sample_source = {'source_size_pc': ['UNIFORM', 1, 10]}
kwargs_sample_macro_fixed = {
    'a4_a': ['GAUSSIAN', 0.0, 0.01], # Gaussian prior on a4
    'a3_a': ['GAUSSIAN', 0.0, 0.005], # Gaussian prior on a3
    'delta_phi_m3': ['UNIFORM', -np.pi/6, np.pi/6]
}

output_path = os.getcwd() + '/example_inference/'
job_index = 1 # unique identifier for output folders
n_keep = 4 # produce this many realizations
tolerance = np.inf # acceptance threshold imposed on the summary statistic tolerance
```
